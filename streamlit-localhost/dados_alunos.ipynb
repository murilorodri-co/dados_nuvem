```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Lista dos estados a serem removidos
estados_remover = ['Balneário Camboriú', 'Cv', 'Du', 'Estocolmo', 'Lb', 'Tg']

# 1. Calcular a porcentagem de cada nível de escolaridade por estado
escolaridade_estado_percentual = df.groupby(['Estado', 'Escolaridade']).size().reset_index(name='Quantidade')
total_por_estado = df.groupby('Estado').size().reset_index(name='Total')

# 2. Mesclar as tabelas para calcular as porcentagens
escolaridade_estado_percentual = pd.merge(escolaridade_estado_percentual, total_por_estado, on='Estado')
escolaridade_estado_percentual['Porcentagem'] = escolaridade_estado_percentual['Quantidade'] / escolaridade_estado_percentual['Total'] * 100

# 3. Filtrar os estados indesejados
escolaridade_estado_percentual = escolaridade_estado_percentual[~escolaridade_estado_percentual['Estado'].isin(estados_remover)]

# 4. Criar uma tabela dinâmica (pivot table)
pivot_df = escolaridade_estado_percentual.pivot(index='Estado', columns='Escolaridade', values='Porcentagem')

# 5. Substituir NaN por 0 para as células vazias
pivot_df = pivot_df.fillna(0)

# 6. Plotar o gráfico de heatmap
plt.figure(figsize=(16, 10))  # Aumentando o tamanho da figura
sns.heatmap(pivot_df, annot=True, fmt='.1f', cmap='coolwarm', cbar_kws={'label': 'Porcentagem de Escolaridade (%)'}, linewidths=0.5)

# 7. Adicionar título e ajustar
plt.title('Distribuição de Níveis de Escolaridade por Estado (%)', fontsize=16)
plt.xlabel('Nível de Escolaridade', fontsize=12)
plt.ylabel('Estado', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Exibir o gráfico
plt.show()
```

```python
import streamlit as st
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Carregar o DataFrame (arquivo Excel)
df = pd.read_excel('alunos_pii_none.xlsx')  # Usando pd.read_excel ao invés de pd.read_csv

# Título do Dashboard
st.title('Dashboard de Escolaridade e Matrícula de Alunos')

# Gráfico 1 - Distribuição de Níveis de Escolaridade
st.subheader('Distribuição de Níveis de Escolaridade (%)')

# Calcular a porcentagem de cada nível de escolaridade
escolaridade_percentual = df['Escolaridade'].value_counts(normalize=True) * 100

# Criar o gráfico de barras
fig, ax = plt.subplots(figsize=(10, 6))
escolaridade_percentual.plot(kind='bar', color=plt.cm.Paired.colors, width=0.8, ax=ax)

# Adicionar título e rótulos
ax.set_title('Distribuição de Níveis de Escolaridade (%)', fontsize=16)
ax.set_xlabel('Nível de Escolaridade', fontsize=12)
ax.set_ylabel('Porcentagem de Alunos (%)', fontsize=12)

# Adicionar as porcentagens acima das barras
for i, v in enumerate(escolaridade_percentual):
    ax.text(i, v + 2, f'{v:.2f}%', ha='center', fontsize=8)

# Exibir o gráfico
st.pyplot(fig)

# Gráfico 2 - Distribuição de Níveis de Escolaridade por Estado (Heatmap)
st.subheader('Distribuição de Níveis de Escolaridade por Estado (%)')

# Lista dos estados a serem removidos
estados_remover = ['Balneário Camboriú', 'Cv', 'Du', 'Estocolmo', 'Lb', 'Tg']

# Calcular a porcentagem de cada nível de escolaridade por estado
escolaridade_estado_percentual = df.groupby(['Estado', 'Escolaridade']).size().reset_index(name='Quantidade')
total_por_estado = df.groupby('Estado').size().reset_index(name='Total')

# Mesclar as tabelas para calcular as porcentagens
escolaridade_estado_percentual = pd.merge(escolaridade_estado_percentual, total_por_estado, on='Estado')
escolaridade_estado_percentual['Porcentagem'] = escolaridade_estado_percentual['Quantidade'] / escolaridade_estado_percentual['Total'] * 100

# Filtrar os estados indesejados
escolaridade_estado_percentual = escolaridade_estado_percentual[~escolaridade_estado_percentual['Estado'].isin(estados_remover)]

# Criar uma tabela dinâmica
pivot_df = escolaridade_estado_percentual.pivot(index='Estado', columns='Escolaridade', values='Porcentagem')

# Substituir NaN por 0
pivot_df = pivot_df.fillna(0)

# Plotar o gráfico de heatmap
fig, ax = plt.subplots(figsize=(16, 10))
sns.heatmap(pivot_df, annot=True, fmt='.1f', cmap='coolwarm', cbar_kws={'label': 'Porcentagem de Escolaridade (%)'}, linewidths=0.5, ax=ax)

# Adicionar título e ajustar
ax.set_title('Distribuição de Níveis de Escolaridade por Estado (%)', fontsize=16)
ax.set_xlabel('Nível de Escolaridade', fontsize=12)
ax.set_ylabel('Estado', fontsize=12)

# Exibir o gráfico
st.pyplot(fig)

# Gráfico 3 - Número de Alunos Matriculados por Faixa Etária
st.subheader('Número de Alunos Matriculados por Faixa Etária')

# Verificar os valores únicos da coluna 'Está Matriculado?' para garantir que os valores sejam 'Sim' e 'Não' ou 'yes' e 'no'
st.write(df['Está Matriculado?'].unique())  # Verificar os valores únicos

# Agrupar os dados por faixa etária e status de matrícula
df_matriculados = df.groupby(['Faixa etária', 'Está Matriculado?']).size().unstack(fill_value=0)

# Remover faixas etárias que não têm alunos
df_matriculados = df_matriculados[df_matriculados.sum(axis=1) > 0]

# Criar o gráfico de barras agrupadas
fig, ax = plt.subplots(figsize=(10, 6))
df_matriculados.plot(kind='bar', color=['#56B4E9', '#D55E00'], ax=ax)

# Adicionar título e rótulos
ax.set_title('Número de Alunos Matriculados por Faixa Etária', fontsize=16)
ax.set_xlabel('Faixa Etária', fontsize=12)
ax.set_ylabel('Quantidade de Alunos', fontsize=12)

# Exibir o gráfico
st.pyplot(fig)

# Exibir a lista com a quantidade de alunos matriculados por faixa etária
st.subheader('Lista de Alunos Matriculados por Faixa Etária')

# Verifique os nomes das colunas, e ajuste para o correto, caso seja necessário
for faixa, (nao, sim) in df_matriculados.iterrows():
    st.write(f'{faixa}: {sim}')  # Variável 'sim' contém o número de matriculados

# Calcular o total de alunos matriculados e não matriculados
# Ajuste os valores para o nome correto dos valores na coluna 'Está Matriculado?'
total_matriculados = df_matriculados['Sim'].sum() if 'Sim' in df_matriculados.columns else df_matriculados.iloc[:, 0].sum()  # Caso o valor seja diferente, ajuste
total_nao_matriculados = df_matriculados['Não'].sum() if 'Não' in df_matriculados.columns else df_matriculados.iloc[:, 1].sum()  # Ajuste conforme necessário

st.write(f"\n**Total de Alunos Matriculados:** {total_matriculados}")
st.write(f"**Total de Alunos Não Matriculados:** {total_nao_matriculados}")
```
